```python
#open sqllite database

import sqlite3
conn = sqlite3.connect('database.sqlite')
c = conn.cursor()   

#get all data
c.execute("SELECT * FROM Player_Attributes")
data = c.fetchall()

import pandas as pd
#convert to dataframe
df = pd.DataFrame(data, columns=[desc[0] for desc in c.description])
print(df.head())
conn.close()
```
