```python
from sklearn.datasets import load_diabetes
X, y = load_diabetes(return_X_y=True)
print(X.shape, y.shape)
```

```python
import numpy as np
print("Mean of each feature:", np.mean(X, axis=0))
print("Standard deviation", np.std(X, axis=0))
print("Maximum values:", np.max(X, axis=0))  
print("Minimum value", np.min(X, axis=0))
print("Median", np.median(X, axis=0))
print("Variance", np.var(X, axis=0))

print("Shape of the dataset:", X.shape)
print("Data type of the features:", X.dtype)
print("First 5 samples:\n", X[:5])
```

```python
import matplotlib.pyplot as plt
plt.boxplot(X)
plt.title("Boxplot of Diabetes Dataset Features")
plt.xlabel("Feature Index")
plt.ylabel("Value")
plt.show()
```

```python
#check if normalization needed
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
print("Original Mean of each feature:", np.mean(X, axis=0))
print("Original Std of each feature:", np.std(X, axis=0)  )
print("Mean after scaling", np.mean(X_scaled, axis=0))
print("Std after scaling", np.std(X_scaled, axis=0))

X_normalized = scaler.fit_transform(X)
print("First 5 samples after normalization:\n", X_normalized[:5])

plt.boxplot(X_normalized)
plt.title("Boxplot of Normalized Diabetes Dataset Features")
plt.xlabel("Feature Index")
plt.ylabel("Value")
plt.show()   
```

```python

import numpy as np
correlation_matrix = np.corrcoef(X, rowvar=False)
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm')
plt.title("Correlation Matrix of Diabetes Dataset Features")
plt.show()  
```
